## Task 4: Unpivot the Data

```python
gdp_new = pd.melt(gdp, id_vars='country', var_name='year',value_name='gdp', ignore_index=True)

print(gdp_new)
```

## Task 5: Standardize Units k, M and B

```python
gdp_new['gdp'] = gdp_new['gdp'].replace({'k': '*1e3', 'M': '*1e6', 'B': '*1e9', np.nan :'0'}, regex=True).map(pd.eval).astype(int)

print(gdp_new)
```

## Task 6: Correct Data Types of Columns

```python
gdp_new.info()
```

```python
gdp_new['year'] = gdp_new['year'].astype(int)
```

## Task 7: Create a Data Cleaning Function

```python
def cleaner(df,value_name,value_name_dtype):
    x = pd.melt(df, id_vars='country', var_name='year', value_name= value_name, ignore_index=True)
    x[value_name] = x[value_name].replace({'k': '*1e3', 'M': '*1e6', 'B': '*1e9', np.nan :'0'}, regex=True).map(pd.eval).astype(value_name_dtype)
    x['year'] = x['year'].astype(int)
    return x
```

```python
pop_new = cleaner(pop, 'pop', 'int')
```

```python
lex_new = cleaner(lex, 'lex', 'float')
```

## Task 8: Merge the Three Datasets

```python
merge1 = gdp_new.merge(pop_new, how='left', left_on=['country', 'year'], right_on=['country','year'])
```

```python
gapminder_df = merge1.merge(lex_new, how='left',left_on=['country', 'year'], right_on=['country','year'])

print(gapminder_df)
```

## Task 9: Drop Null Rows

```python
gapminder_df.dropna(inplace=True)
```

```python
gapminder_df.reset_index(drop=True, inplace=True)
```

```python
gapminder_df['gdp'] = gapminder_df['gdp'].astype(int)
gapminder_df['pop'] = gapminder_df['pop'].astype(int)

print(gapminder_df)
```

## Task 10: Generate Continent Column

```python
gapminder_df['country_code'] = gapminder_df['country'].apply(lambda x: pc.country_name_to_country_alpha2(x, cn_name_format="default"))
```

```python
gapminder_df[gapminder_df['country_code']=='TL'] = gapminder_df[gapminder_df['country_code']=='TL'].assign(country_code='TP')
```

```python
gapminder_df['continent_code'] = gapminder_df['country_code'].apply(lambda x: pc.country_alpha2_to_continent_code(x))
```

```python
gapminder_df['continent'] = gapminder_df["continent_code"].apply(lambda x: pc.convert_continent_code_to_continent_name(x))
```

```python
gapminder_df.drop(columns=['country_code','continent_code'], inplace=True)

print(gapminder_df)
```

## Task 11: Filter Data by Years

```python
df = gapminder_df[(gapminder_df['year'] > 1899) & (gapminder_df['year'] < 2024)]
print(df)
```

## Task 12: Plot the Data

```python
fig = px.scatter(df, x='gdp', y='lex', animation_frame='year', animation_group='country',
       size='pop', color='continent', hover_name='country', log_x = True, size_max=100,
      range_x=[500,200000], range_y=[0,90], labels=dict(gdp="GDP Per Capita ($)",
        lex="Life Expectancy (years)", continent="Continents"),
        title = 'GDP Per Capita, Life Expectancy & Population (1900-2023)')
fig.update_layout(title_x=0.5,
              xaxis=dict(showgrid=False, tickmode = 'array',
                tickvals = [500, 1000, 10000, 100000, 200000],
                ticktext = [500, 1000, '10K', '100K','200K']),
              yaxis=dict(showgrid=False))

fig.show()
```

## Task 13: Save Chart as GIF

```python
# Create animated GIF from Plotly figure
import time
from PIL import Image

print("Creating animated GIF... This may take a moment.")

# Initialize empty list for frames
frames = []

# Create frames from each animation frame
for i, frame in enumerate(fig.frames):
    print(f"Processing frame {i+1}/{len(fig.frames)}")
    
    # Update figure with current frame data
    fig_copy = fig
    fig_copy.update(data=frame.data)
    
    # Update animation slider position
    if fig_copy.layout.sliders:
        fig_copy.layout.sliders[0].update(active=i)
    
    # Convert to PIL Image and add to frames list
    img_bytes = fig_copy.to_image(format="png", width=800, height=600)
    pil_img = Image.open(io.BytesIO(img_bytes))
    frames.append(pil_img)

# Save as animated GIF
output_path = r"G:\PROJECTS\TIME SERIES\Time Series Analysis using Python and Plotly\animated_chart.gif"

print(f"Saving GIF with {len(frames)} frames...")
frames[0].save(
    output_path,
    save_all=True,
    append_images=frames[1:],
    optimize=True,
    duration=200,  # Duration between frames in milliseconds
    loop=0  # 0 means infinite loop
)

print(f"✅ Animated GIF saved successfully to: {output_path}")
print(f"📊 Total frames: {len(frames)}")
print(f"🎬 Animation duration: {len(frames) * 0.2:.1f} seconds")
```
